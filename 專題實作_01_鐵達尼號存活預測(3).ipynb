```python
import pandas as pd
df_train = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
df_train

```

# **資料清理 (null欄位)**

##** 資料分析是找趨勢的分析 **
#### 像是name/passengerid這種資料就不適合

```python
#利用info()找出null欄位
df_train.info()

#類別型:sex , ticket, cabin, embarked
#數值型:pclass, age, sibsp, parch, fare
```

## 從上述資料來看 age/Cabin/Embarked 有空值

```python
#不考慮準確度:暴力解法
#將空值都補上
df_train['Age']=df_train['Age'].fillna(0)
df_train['Embarked']=df_train['Embarked'].fillna('None')

#Cabin 空值數量太多不考慮 直接捨棄
df_train.info()
```

# **資料型態轉換**

```python
print(df_train['Sex'])
print(df_train['Sex'].value_counts())
```

```python
df_train['Sex'] = df_train['Sex'].replace({'male': 0, 'female': 1})
print(df_train['Sex'])
```

```python

df_Embarked_ohe = pd.get_dummies(df_train['Embarked'], prefix='Embarked_')
# 將結果布林值轉成數值
df_Embarked_ohe = df_Embarked_ohe.astype(int)

print(df_Embarked_ohe)
```

```python
# 將轉成數值型態的sex,embarked處理回原本資料
df_train = pd.concat([df_train, df_Embarked_ohe], axis=1)

print(df_train)
```

# **模型分析**

```python
#目前資料的欄位
df_train.columns
```

```python
columns_X = ['Pclass', 'Sex', 'Age', 'SibSp','Parch', 'Fare', 'Embarked__C','Embarked__None', 'Embarked__Q', 'Embarked__S']
columns_y = ['Survived']


train_X = df_train[columns_X]
train_y = df_train[columns_y]

#print(df_train['Embarked'])

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

#cross_val_score()
log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(),cv=5,scoring='accuracy')
print(scores)
```
