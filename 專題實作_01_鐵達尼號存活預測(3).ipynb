```python
import pandas as pd
df_train = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
df_train

```

# **資料清理 (null欄位)**

##** 資料分析是找趨勢的分析 **
#### 像是name/passengerid這種資料就不適合

```python
#利用info()找出null欄位
df_train.info()

#類別型:sex , ticket, cabin, embarked
#數值型:pclass, age, sibsp, parch, fare
```

## 從上述資料來看 age/Cabin/Embarked 有空值

```python
#不考慮準確度:暴力解法
#將空值都補上
df_train['Age']=df_train['Age'].fillna(0)
df_train['Embarked']=df_train['Embarked'].fillna('None')

#Cabin 空值數量太多不考慮 直接捨棄
df_train.info()
```

# **資料型態轉換**

```python
print(df_train['Sex'])
print(df_train['Sex'].value_counts())
```

```python
df_train['Sex'] = df_train['Sex'].replace({'male': 0, 'female': 1})
print(df_train['Sex'])
```

```python

df_Embarked_ohe = pd.get_dummies(df_train['Embarked'], prefix='Embarked_')
# 將結果布林值轉成數值
df_Embarked_ohe = df_Embarked_ohe.astype(int)

print(df_Embarked_ohe)
```

```python
# 將轉成數值型態的sex,embarked處理回原本資料
df_train = pd.concat([df_train, df_Embarked_ohe], axis=1)

print(df_train)
```

# **模型分析**

```python
#目前資料的欄位
df_train.columns
```

```python
columns_X = ['Pclass', 'Sex', 'Age', 'SibSp','Parch', 'Fare', 'Embarked__C','Embarked__None', 'Embarked__Q', 'Embarked__S']
columns_y = ['Survived']


train_X = df_train[columns_X]
train_y = df_train[columns_y]

#print(df_train['Embarked'])

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

#cross_val_score()
log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(),cv=5,scoring='accuracy')
print(scores)
```

# **Task#04 特徵工程**

```python
#將乘客頭銜分類
df_train['Title Row'] = df_train['Name'].map( lambda name: name.split( ',' )[1].split( '.' )[0].strip() )

title_encoding = {
  "Capt":       "Officer",
  "Col":        "Officer",
  "Major":      "Officer",
  "Jonkheer":   "Royalty",
  "Don":        "Royalty",
  "Sir" :       "Royalty",
  "Dr":         "Officer",
  "Rev":        "Officer",
  "the Countess":"Royalty",
  "Dona":       "Royalty",
  "Mme":        "Mrs",
  "Mlle":       "Miss",
  "Ms":         "Mrs",
  "Mr" :        "Mr",
  "Mrs" :       "Mrs",
  "Miss" :      "Miss",
  "Master" :    "Master",
  "Lady" :      "Royalty"
}

df_train['Title'] = df_train['Title Row'].map(title_encoding)
df_train['Title']
```

```python
#pd.get_dummies:將類別型態資料轉成0或1表示
df_Title_ohe = pd.get_dummies(df_train['Title'], prefix='Title_')
#將原始的true/false轉成數值型態
df_Title_ohe = df_Title_ohe.astype(int)
df_Title_ohe
```

```python
#將乘客分類後的結果加回資料中
df_train = df_train.join(df_Title_ohe)
df_train
```

### **Parch(兄弟姐妹/配偶) + SibSp(父母/兒童) = 家庭人數 與生存之相關性**

```python
#兄弟姊妹+父母+自己
df_train['FamilySize'] = df_train['Parch'] + df_train['SibSp'] + 1
df_train['FamilySize']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr = df_train[['Survived', 'Age', 'Fare', 'FamilySize']].corr()
cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
sns.heatmap(
        corr,
        cmap = cmap,
        square=True,
        cbar_kws={ 'shrink' : .9 },
        ax = plt.subplots( figsize =( 12 , 10 ) )[1],
        annot = True,
        annot_kws = { 'fontsize' : 12 }
)
```

```python
def plot_distribution( df , var , target , **kwargs ):
    row = kwargs.get( 'row' , None )
    col = kwargs.get( 'col' , None )
    facet = sns.FacetGrid( df , hue=target , aspect=4 , row = row , col = col )
    facet.map( sns.kdeplot , var , fill=True )
    facet.set( xlim=( 0 , df[ var ].max() ) )
    facet.add_legend()
```

```python
plot_distribution( df_train , var = 'FamilySize' , target = 'Survived' )
plt.show()
```

```python
# 將家庭人數分成區間可能較能看出趨勢
#如果人數=1 則表格中的Family_Single會顯示1(s : 1)
df_train['Family_Single'] = df_train[ 'FamilySize' ].map( lambda s : 1 if s == 1 else 0 )
df_train['Family_Small']  = df_train[ 'FamilySize' ].map( lambda s : 1 if 2 <= s <= 4 else 0 )
df_train['Family_Large']  = df_train[ 'FamilySize' ].map( lambda s : 1 if 5 <= s else 0 )

#針對'FamilySize'的人數
def FamilySizeTranform(s):
  if s >= 1 and s <= 3:
    return 3
  if s >= 4 and s <= 6:
    return 2
  return 1

df_train['FamilySizeEncoding'] = df_train[ 'FamilySize' ].map(FamilySizeTranform)

df_train
```

## **模型分析**

**原始資料**

```python
columns_X = [
       'Pclass', 'Sex', 'Age', 'Fare',
       'SibSp', 'Parch',
       'Embarked__C', 'Embarked__None', 'Embarked__Q', 'Embarked__S',
]
columns_y = ['Survived']

train_X = df_train[columns_X]
train_y = df_train[columns_y]

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(),cv=5,scoring='accuracy')
print(scores)
print(scores.mean())
```
